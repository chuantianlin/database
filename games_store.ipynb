# import data from CSV TO TABLE

```python
def insert_data(TableName,File_address):
    cursor = conn.cursor()
    if "/" in File_address:
        "//".join(File_address.split('/'))
    data = pd.read_csv (File_address)   
    df = pd.DataFrame(data)
    
    value=df.values.tolist()
    for i in range(len(value)):
       
        for j in range(len(value[i])):
            try:
                if math.isnan(float(value[i][j])):
                    value[i][j]="NULL"
                
                else:
                    value[i][j]=("'% s'" % value[i][j])
            except:
                    value[i][j]=("'% s'" % value[i][j])
        query = 'insert into '+TableName+' values ({0})'
        query = query.format(','.join(value[i]))
        cursor.execute(query)
           
    cursor.execute('commit')
   
insert_data('products','products.csv')
```

# Populate data from TABLE

```python
def pop_data(TableName):
    cursor = conn.cursor()
    query = "SELECT * FROM "+TableName
    cursor.execute(query)
    records = [str(record)+'\n' for record in cursor]
    for i in records:
        print(i)
pop_data("products")
```

```python
def pop_data_condition(TableName,*Boolean_operator,**condition):
    cursor = conn.cursor()
    query1 = "SELECT * FROM "+TableName
    string=[]
    condition=list(condition.items())
    operator=[] if not Boolean_operator else list (Boolean_operator)
    
    if len(operator)>=len(condition) or len(operator)!=len(condition)-1:
        print('syntax error,please match the number of binary oprtator and the conditions thanks!')
        return
    index=[i for  i in range(len(condition))]
    if len(condition) >= 1:
            for index,V in zip(index,condition):
                string.append('%s="%s"'%(list(V)[0],list(V)[1])+('' if  index >=len(operator) else operator[index]+' '))
        
            query1=query1+" where "+ (''.join(string) if len(string)>=2 else string[0])
            print(query1)
    cursor.execute(query1)
    records = [str(record)+'\n' for record in cursor]
    if records:
        for i in records:
            print(i)
    else:
        print(None)
pop_data_condition("products",'and',name="Grand Theft auto",price=120)
```
